# Qualtitative Data

## Education Feature

```python
df_edu_groub = df_main.groupby(['Default', 'Education'])['LoanID'].count().unstack()
titles       = df_main['Education'].value_counts().index
cols         = len(titles)
col_no       = 0
```

```python
fig = mpp.figure(figsize=(12, 6))

for n in titles:
    ax1 = mpp.subplot2grid((1, cols), (0, col_no))
    mpp.pie(df_edu_groub[n], colors = ('#ffa5a5', '#a5ffa5'), autopct = '%.2f', labels = [0, 1], explode = [0.0, 0.1])
    mpp.title(n)
    col_no += 1
    
mpp.show()
```

## HasMortgage, HasDependents and HasCoSigner Features

```python
df_mor_groub = df_main.groupby(['Default', 'HasMortgage'])['LoanID'].count().unstack()
df_dep_groub = df_main.groupby(['Default', 'HasDependents'])['LoanID'].count().unstack()
df_cos_groub = df_main.groupby(['Default', 'HasCoSigner'])['LoanID'].count().unstack()
titles       = ['No', 'Yes']
features     = ['HasMortgage', 'HasDependents', 'HasCoSigner']
cols         = 3
rows         = 2
col_no       = 0
row_no       = 0
```

```python
fig = mpp.figure(figsize=(12, 6))

for n in titles:
    for d in [df_mor_groub, df_dep_groub, df_cos_groub]:
        ax1 = mpp.subplot2grid((rows, cols), (row_no, col_no))
        mpp.pie(d[n], colors = ('#ffa5a5', '#a5ffa5'), autopct = '%.2f', labels = [0, 1], explode = [0.0, 0.1])
        mpp.title(f'{features[row_no]} - {n}')
        col_no += 1
    col_no = 0
    row_no += 1
    
mpp.show()
```

```python
quant_df = df_main[['Age', 'Income', 'LoanAmount', 'CreditScore', 'MonthsEmployed', 'NumCreditLines', 'LoanTerm', \
                    'InterestRate', 'DTIRatio', 'Default']]
```

```python
quant_df.describe()
```

```python
df_new['Age']            = quant_df['Age'].apply(ConvertAgeValues)
df_new['Income']         = quant_df['Income'].apply(ConvertIncomeValues)
df_new['CreditScore']    = quant_df['CreditScore'].apply(ConvertCreditScoreValues)
df_new['LoanAmount']     = quant_df['LoanAmount'].apply(ConvertLoanAmountValues)
df_new['LoanTerm']       = quant_df['LoanTerm'].apply(ConvertLoanTermValues)
df_new['MonthsEmployed'] = quant_df['MonthsEmployed'].apply(ConvertMonthsEmployedValues)
df_new['InterestRate']   = quant_df['InterestRate'].apply(ConvertInterestRateValues)
df_new['DTIRatio']       = quant_df['DTIRatio'].apply(ConvertDTIRatioValues)
```

```python
df_new.head()
```

```python
new_corr = df_new.corr()
```

```python
correlations = pd.DataFrame({'Old_Corr' : df_corr['Default'].values,
                             'New_Corr' : new_corr['Default'].values}, index = df_corr.index)
```

```python
mpp.figure(figsize=(4, 6)), 
sb.heatmap(data=correlations, fmt='.3f', annot=True, cmap='hot')
mpp.show()
```
